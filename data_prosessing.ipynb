```python
# df_no_nan = new_df.dropna()
# import random
import numpy as np
# Filter rows without NaN values or empty strings
filtered_df = new_df.dropna().replace('', np.nan).dropna()

# Randomly select 50 rows
no_nan_df_50 = filtered_df.sample(n=50, random_state=42)

# Display the randomly selected rows
len(no_nan_df_50)
```

```python
import tiktoken

def calculate_average_token_count(df):
    total_tokens = 0
    total_rows = len(df)

    for index, row in no_nan_df_50.iterrows():
        text = row['Description']  # Replace 'Description' with the column containing the text
        encoding = tiktoken.get_encoding("cl100k_base")
        encoding = tiktoken.encoding_for_model("gpt-3.5-turbo")
        total_tokens = len(encoding.encode(str(text)))

    average_token_count = total_tokens / total_rows
    return average_token_count

average_tokens = calculate_average_token_count(df)
print(f"Average number of tokens in a row: {average_tokens * 1000:.2f}")
```
